## คำถาม

1. Static Method คืออะไร


เมธอดที่ถูกประกาศด้วยคีย์เวิร์ด static ภายในคลาสในภาษา C# ซึ่งหมายความว่าเมธอดนี้สามารถเรียกใช้งานได้โดยไม่ต้องสร้างอ็อบเจ็กต์ของคลาสนั้น ๆ

2. ประกาศ static method ในภาษา C# อย่างไร


ใช้คีย์เวิร์ด static ก่อนชื่อเมธอดที่ต้องการประกาศให้เป็น static method โดยไม่ต้องสร้างอ็อบเจ็กต์ของคลาสนั้น ๆ ก็สามารถเรียกใช้เมธอดนี้ได้

3. เรียกใช้ static method อย่างไร


เรียกใช้จากชื่อคลาส โดยตรง เช่น MathOperations.Add(5, 3); โดยไม่จำเป็นต้องสร้างอ็อบเจ็กต์ เพราะ static method เป็นของคลาส ไม่ใช่อ็อบเจ็กต์

4. static method สามารถเข้าถึง instance members (fields, methods) ได้หรือไม่


ไม่สามารถเข้าถึง instance members ของคลาสได้โดยตรง เนื่องจาก static method ถูกออกแบบให้ทำงานโดยไม่ต้องพึ่งพาข้อมูลของอ็อบเจ็กต์หรือ instance ของคลาสนั้น ๆ

5. static method เหมาะกับการใช้งานแบบไหน


เหมาะสำหรับการใช้งานในกรณีที่ไม่จำเป็นต้องใช้ข้อมูลจากอ็อบเจ็กต์ของคลาสนั้น ๆ หรือเมธอดที่ทำงานเป็นฟังก์ชันทั่วไปที่ไม่ได้ขึ้นกับสถานะของอ็อบเจ็กต์ใดๆ

6. ข้อดีของการใช้ static method คืออะไรบ้าง


ไม่จำเป็นต้องสร้างอ็อบเจ็กต์ เพราะ Static method สามารถถูกเรียกใช้ได้โดยตรงจากชื่อคลาส โดยไม่ต้องสร้างอ็อบเจ็กต์ของคลาสนั้นๆ 

Static method สามารถเข้าถึงและจัดการกับ static fields หรือข้อมูลที่ถูกประกาศเป็น static ได้โดยตรง ทำให้เหมาะสำหรับการจัดการข้อมูลที่ใช้ร่วมกันในหลายๆ อ็อบเจ็กต์

7. ข้อเสียของการใช้ static method คืออะไรบ้าง


Static method ไม่สามารถเข้าถึงหรือใช้งาน instance members ได้ เนื่องจาก static method ไม่เกี่ยวข้องกับการสร้างอ็อบเจ็กต์ในคลาสนั้นๆ หากต้องการใช้งาน instance members จะต้องสร้างอ็อบเจ็กต์ขึ้นมาก่อน

8. จงยกตัวอย่าง static method ที่ใช้บ่อยใน Library มาตรฐานของภาษา C# พร้อมทั้งยกตัวอย่างการใช้งาน

Math.Abs() เป็น static method ที่ใช้ในการหาค่าสัมบูรณ์ (absolute value) ของตัวเลข

````
using System;

class Program
{
    static void Main()
    {
        int negativeNumber = -25;
        int absoluteValue = Math.Abs(negativeNumber);
        
        Console.WriteLine($"The absolute value of {negativeNumber} is {absoluteValue}");
    }
}
````
