## คำถาม

1. Static field และ static method ต่างจาก instance field และ instance method อย่างไร

Static Field : เป็นฟิลด์ที่เป็นสมาชิกของคลาส และสามารถเข้าถึงได้โดยไม่ต้องสร้างอ็อบเจ็กต์

Instance Field : เป็นฟิลด์ที่เป็นสมาชิกของอ็อบเจ็กต์แต่ละตัว ต้องสร้างอ็อบเจ็กต์ก่อนถึงจะเข้าถึงได้

Static Method : เป็นเมธอดที่ใช้ร่วมกันระหว่างทุกอ็อบเจ็กต์ของคลาส สามารถเรียกใช้ได้จากชื่อคลาสโดยตรง

Instance Method : เป็นเมธอดที่ทำงานกับอ็อบเจ็กต์ ต้องสร้างอ็อบเจ็กต์ก่อนถึงจะเรียกใช้ได้

2. เหตุใดเราจึงสามารถเข้าถึง Person.name และ Person.GetName() ได้โดยตรงผ่านชื่อ class โดยไม่ต้องสร้าง object

Person.name และ Person.GetName() สามารถเข้าถึงได้โดยไม่ต้องสร้างอ็อบเจ็กต์ของ Person เพราะเป็น static members ที่เชื่อมโยงกับคลาส Person โดยตรง


3. Static class มีข้อจำกัดอะไรบ้าง


1. ไม่สามารถสร้างอ็อบเจ็กต์ ของคลาส

2. ทุกฟิลด์และเมธอดต้องเป็น static

3. ไม่สามารถสืบทอด คลาส static ได้

4. ไม่สามารถประกาศเป็น abstract หรือ sealed

5. ไม่สามารถมี constructor ที่ไม่ใช่ static

6. ไม่สามารถใช้ instance members หรือ this ภายใน static methods.1

4. เราไม่สามารถสร้าง instance ของ static class ได้เพราะอะไร

Static class ถูกออกแบบให้เป็นคลาสที่ไม่มีสถานะอ็อบเจ็กต์ และมีฟังก์ชันที่ใช้ร่วมกันจากชื่อคลาส ซึ่งไม่มีตัวสร้างที่สามารถสร้างอ็อบเจ็กต์จาก static class

การสร้างอ็อบเจ็กต์จาก static class จะไม่มีความหมาย เพราะมันออกแบบให้ใช้งานโดยตรงจากชื่อคลาส

5. ประโยชน์ของการใช้ static class คืออะไร  มีสถานการณ์ใดบ้างที่เราควรใช้ static class

มีประโยชน์ในการจัดการฟังก์ชันหรือข้อมูลที่ไม่เกี่ยวข้องกับอ็อบเจ็กต์ที่สร้างขึ้นจากคลาสนั้นๆ เช่น ฟังก์ชันคำนวณหรือค่าคงที่ที่ใช้ร่วมกันในหลายที่ในโปรแกรม 

โดยไม่จำเป็นต้องสร้างอ็อบเจ็กต์ใหม่ทุกครั้งที่ต้องการใช้งาน ฟังก์ชันและข้อมูลใน static class จะเป็น static หมายความว่าไม่ขึ้นอยู่กับสถานะของอ็อบเจ็กต์ใดๆ 

และสามารถเข้าถึงได้โดยตรงจากชื่อคลาส ตัวอย่างเช่น ต้องการสร้างฟังก์ชันที่คำนวณหรือแปลงข้อมูลที่ไม่เกี่ยวข้องกับสถานะของอ็อบเจ็กต์ 

สามารถใส่เมธอดเหล่านั้นไว้ใน static class เพื่อให้สามารถเรียกใช้ได้ง่ายจากทุกที่ในโปรแกรมโดยไม่ต้องสร้างอ็อบเจ็กต์ใหม่
