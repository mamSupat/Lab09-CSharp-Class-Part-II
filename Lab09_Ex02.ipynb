## คำถาม

1. เมื่อกำหนดให้สมาชิกในคลาสให้มี Access Modifier ดังต่อไปนี้ จะสามารถเข้าถึงจากที่ไหนได้บ้าง


1.1 public

สมาชิกสามารถเข้าถึงได้จากทุกที่ในโปรแกรม ไม่ว่าจะเป็นจากภายในคลาสเดียวกัน, คลาสอื่นๆ, หรือแม้กระทั่งโปรเจ็กต์อื่นๆ ที่มีการอ้างถึงคลาสนี้

1.2 private

สามารถเข้าถึงได้เฉพาะภายในคลาสหรือโค้ดที่ประกาศสมาชิกนั้นๆ เท่านั้น ไม่สามารถเข้าถึงจากภายนอกได้

1.3 protected

สมาชิกสามารถเข้าถึงได้จากคลาสที่ประกาศสมาชิกนั้นๆ และคลาสที่สืบทอด (inherit) คลาสนั้น (subclasses) เท่านั้น 
ไม่สามารถเข้าถึงจากภายนอกคลาสหรือโปรเจ็กต์อื่นได้

1.4 internal

สมาชิกสามารถเข้าถึงได้จากภายในโปรเจ็กต์เดียวกันเท่านั้น (หรือ assembly เดียวกัน) ไม่สามารถเข้าถึงจากโปรเจ็กต์หรือ assembly อื่นได้

1.5 protected internal

สมาชิกสามารถเข้าถึงได้จากภายในโปรเจ็กต์เดียวกันและจากคลาสที่สืบทอดจากคลาสนั้นๆ แม้จะอยู่ในโปรเจ็กต์ต่างกัน
