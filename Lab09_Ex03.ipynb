## คำถาม

1. ในภาษา C# นั้น Instance members และ static members  แตกต่างกันอย่างไร

Instance Members : เกี่ยวข้องกับอ็อบเจ็กต์ ต้องสร้างอ็อบเจ็กต์ก่อนเข้าถึง ค่าของสมาชิกนี้จะแตกต่างกันไปในแต่ละอ็อบเจ็กต์

Static Members : เกี่ยวข้องกับคลาสและมีค่าเดียวกันสำหรับทุกอ็อบเจ็กต์ที่สร้างจากคลาสนั้น ไม่ต้องสร้างอ็อบเจ็กต์เพื่อเข้าถึงสมาชิกเหล่านี้

2. เข้าถึง Instance members และ static members มีหลักการที่แตกต่างกันอย่างไร

Instance Members : ต้องสร้างอ็อบเจ็กต์ของคลาสก่อนจึงจะสามารถเข้าถึงได้

Static Members : สามารถเข้าถึงได้ผ่านชื่อคลาสโดยตรงโดยไม่จำเป็นต้องสร้างอ็อบเจ็กต์

3. Instance members และ static members เหมาะกับการใช้งานแบบไหน

Instance Members : ใช้เมื่อคุณต้องการให้ข้อมูลหรือฟังก์ชันแยกกันระหว่างอ็อบเจ็กต์แต่ละตัว และแต่ละอ็อบเจ็กต์อาจมีค่าที่แตกต่างกัน

Static Members : ใช้เมื่อข้อมูลหรือฟังก์ชันมีค่าร่วมกันสำหรับทุกอ็อบเจ็กต์ในคลาส

4. Instance members และ static members ใช้หน่วยความจำต่างกันอย่างไร

Instance Members :

ถูกเก็บในหน่วยความจำแยกสำหรับแต่ละอ็อบเจ็กต์

การใช้หน่วยความจำเพิ่มขึ้นตามจำนวนอ็อบเจ็กต์ที่สร้าง

ข้อมูลใน instance members จะแตกต่างกันในแต่ละอ็อบเจ็กต์

Static Members :

ถูกเก็บในหน่วยความจำที่แชร์ร่วมกันสำหรับทุกอ็อบเจ็กต์ของคลาส

การใช้หน่วยความจำจะไม่เพิ่มขึ้นตามจำนวนอ็อบเจ็กต์ เพราะ static members ถูกสร้างเพียงครั้งเดียว

ข้อมูลใน static members จะเหมือนกันสำหรับทุกอ็อบเจ็กต์ของคลาส

5. Instance members และ static members มีอายุการใช้งานต่างกันอย่างไร

Instance Members : อายุการใช้งานของ instance members จะสัมพันธ์กับอายุของอ็อบเจ็กต์ที่สร้างขึ้นจากคลาสนั้น ๆ และจะถูกทำลายเมื่ออ็อบเจ็กต์นั้นถูกทำลายหรือไม่ถูกอ้างอิงอีกต่อไป

Static Members : อายุการใช้งานของ static members จะสัมพันธ์กับอายุของคลาสที่ประกาศ static member นั้น ๆ และจะมีอายุการใช้งานจนกระทั่งโปรแกรมสิ้นสุดหรือคลาสถูกโหลดใหม่

6. Instance members และ static members สามารถใช้งานร่วมกันได้หรือไม่

Instance members และ static members สามารถใช้งานร่วมกันได้ภายในคลาสเดียวกัน แต่ต้องคำนึงถึงวิธีการเข้าถึงที่แตกต่างกัน

Static members สามารถเข้าถึง instance members ได้แต่ต้องสร้างอ็อบเจ็กต์

Instance members สามารถเข้าถึง static members ได้โดยไม่จำเป็นต้องสร้างอ็อบเจ็กต์

Static methods ต้องสร้างอ็อบเจ็กต์เพื่อเข้าถึง instance members เนื่องจากไม่สามารถเข้าถึง instance members โดยตรงได้
